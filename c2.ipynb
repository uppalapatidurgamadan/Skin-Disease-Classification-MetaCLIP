```python
import os
import numpy as np
import scipy.io
from sklearn.preprocessing import *
from numpy import mean
from numpy import std
from sklearn.model_selection import KFold
from sklearn.datasets import make_blobs
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import RidgeClassifier
from sklearn.linear_model import RidgeClassifierCV
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import StratifiedKFold
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.impute import SimpleImputer
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.dummy import DummyClassifier
from lightgbm import LGBMClassifier #Added this line
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import NearestCentroid

acc=[]
#pca=PCA(n_components=300)
#3.OFMM10.mat,svc 3, cv..5...rand1..93.4
#20 the best
entries=os.listdir('/content/gdrive/MyDrive/skinmat1/') # Changed this line
for i in range(2, 3):
    # Construct the full file path by joining the directory and filename
    file_path = os.path.join('/content/gdrive/MyDrive/skinmat1/', entries[i])

    mat1 = scipy.io.loadmat(file_path) # Load data using the full file path
    X1 = mat1.get('final')
    X2 = np.array(X1)
    num_rows, num_cols = X2.shape
    featcol=num_cols-1
    Xa=X2[0:num_rows, 0:featcol]
    y=X2[0:num_rows, featcol]
    scaler = MinMaxScaler()
    scaler.fit(Xa)
    Xnew=scaler.transform(Xa)

    # Impute NaN values with the mean of each column
    imputer = SimpleImputer(strategy='mean')
    Xnew = imputer.fit_transform(Xnew)

    cv= RepeatedStratifiedKFold(n_splits=10, n_repeats=1,  random_state=1)
    #cv = KFold(n_splits=5, shuffle=True, random_state=1)
# create model
    #model=LGBMclassifier()
    #model=LinearDiscriminantAnalysis()
    #model=LogisticRegression()
    #model=RidgeClassifierCV()
    #model=NearestCentroid()
    model=SVC()
    #model =LinearDiscriminantAnalysis()
    #balscore=valscore(Xa,y)
# evaluate model
    scores = cross_val_score(model, Xnew, y, scoring='balanced_accuracy', cv=cv, n_jobs=-1)
# report performance
    print('ModelNo:',i,'-','Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
    acc.append(mean(scores))
    #acc.apend(balscore)
    #print('ModelNo:',i,'-',balscore)
```
